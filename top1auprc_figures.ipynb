
# Function: `double_axis_1`
This function creates a dual-axis plot comparing `Top1AUPR` and `Top1AUROC` metrics against the number of proteins remaining after the threshold filter. The first axis (primary) plots `Top1AUPR`, while the second axis (secondary) plots `Top1AUROC`. The y-axis scales for both metrics are matched.

```python
double_axis_1()
```

# Function: `double_axis_2`

This function creates a dual-axis plot similar to `double_axis_1`, but maintains the original scale for the secondary axis (`Top1AUROC`). The y-axis scales for both metrics are not matched, scaled based on the min & max for each metric.

```python
double_axis_2()
```

# Phylogenetic composite score

```python
mutual = pd.read_csv('mutual_top.composite_score.tsv', sep='\t')
```

```python
comp = pd.read_csv('top_5_rankings_composite_score.csv')
```

```python
comp
```

```python
comp.rename(columns={'Gene':'Gene1', 'Top_1':'Gene2'},inplace=True)
```

```python
comp[comp['Gene1']=='AAMP']
```

```python
mutual_score_df = mutual.merge(comp, on = ['Gene1', 'Gene2'])[['Gene1', 'Gene2', 'Score_1']]
```

```python
mutual_score_df1_2 = mutual_score_df.copy()
mutual_score_df2_1 = mutual_score_df.copy().rename(columns={'Gene1':'Gene2', 'Gene2':'Gene1'})
mutual_score_df1_2
```

```python
mutual_score_df2_1
```

```python
#cmplx = pd.read_pickle('complex_label.pkl')
```

```python
#cmplx[['Gene1','Gene2']] = cmplx['GeneAB'].str.split('_', expand = True)
```

```python
cmplx
```

```python
c1 = cmplx.merge(mutual_score_df1_2, on = ['Gene1', 'Gene2'])
```

```python
c2 = cmplx.merge(mutual_score_df2_1, on = ['Gene1', 'Gene2'])

c2
```

```python
c1
```

```python
def remove_matching_rows(df, col1, col2):
    """
    Print and remove rows where df[col1] == df[col2].

    Parameters:
    df (pd.DataFrame): The input dataframe.
    col1 (str): Name of the first column.
    col2 (str): Name of the second column.

    Returns:
    pd.DataFrame: A new dataframe with matching rows removed.
    """
    matching_rows = df[df[col1] == df[col2]]
    print("Matching rows:")
    print(matching_rows)

    new_df = df[df[col1] != df[col2]].copy()
    return new_df
```

```python
c1_ = remove_matching_rows(c1, 'Gene1', 'Gene2')
```

```python
c2_ = remove_matching_rows(c2, 'Gene1', 'Gene2')
```

```python
c_ = pd.concat([c1_, c2_])
```

```python
c_
```

```python
c_['Same_Complex'].value_counts()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import precision_recall_curve, average_precision_score

def plot_pr_curve(df, score_col='Score_1', label_col='Same_Complex'):
    y_score = df[score_col]
    y_true = df[label_col]

    precision, recall, _ = precision_recall_curve(y_true, y_score)
    ap = average_precision_score(y_true, y_score)

    plt.figure(figsize=(8, 6))
    plt.plot(recall, precision, label=f"AP = {ap:.2f}")
    plt.xlabel("Recall")
    plt.ylabel("Precision")
    plt.title("Precision-Recall Curve")
    plt.legend(loc="lower left")
    plt.tight_layout()
    plt.savefig("pr_curve.png", dpi=300)
    plt.show()
    
def plot_roc_curve(df, score_col='Score_1', label_col='Same_Complex'):
    y_score = df[score_col]
    y_true = df[label_col]

    fpr, tpr, _ = roc_curve(y_true, y_score)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.2f}")
    plt.plot([0, 1], [0, 1], 'k--', lw=1)
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve")
    plt.legend(loc="lower right")
    plt.tight_layout()
    plt.savefig("roc_curve.png", dpi=300)
    plt.show()

```

```python
plot_roc_curve(c_)
```

```python
plot_pr_curve(c_)
```

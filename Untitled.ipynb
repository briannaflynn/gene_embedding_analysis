```python
import pandas as pd
import numpy as np
from itertools import combinations, product
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned CSV version of the phylogenetic profile
phylo_df = pd.read_csv("phylo_profiles.csv", index_col=0, low_memory = False, skiprows= 4)
phylo_df.index = phylo_df.index.str.strip()

# Load positive and negative gene sets
positive_genes = pd.read_csv("positive_set.csv", encoding='utf-16')["Gene Name"].tolist()
```

```python
# Filter genes that are present in the phylogenetic matrix
positive_genes_filtered = [g for g in positive_genes if g in phylo_df['Gene Name'].to_list()]
negative_genes_filtered = [g for g in negative_genes if g in phylo_df['Gene Name'].to_list()]

print(f"Retained {len(positive_genes_filtered)} positive genes and {len(negative_genes_filtered)} negative genes")

# Generate gene pairs
from itertools import combinations, product

pp_pairs = list(combinations(positive_genes_filtered, 2))
pn_pairs = list(product(positive_genes_filtered, negative_genes_filtered))
nn_pairs = list(combinations(negative_genes_filtered, 2))

# Optional: sample 1000 of each to keep compute fast
import numpy as np
np.random.seed(42)

pp_sample = np.random.choice(len(pp_pairs), size=min(1000, len(pp_pairs)), replace=False)
pn_sample = np.random.choice(len(pn_pairs), size=min(1000, len(pn_pairs)), replace=False)
nn_sample = np.random.choice(len(nn_pairs), size=min(1000, len(nn_pairs)), replace=False)
pp_sample
def phylo_similarity(g1, g2, df):
    try:
        # Find the row for each gene
        row1 = df[df['Gene Name'] == g1]
        row2 = df[df['Gene Name'] == g2]
        
        if row1.empty or row2.empty:
            return np.nan
        
        # Select only numeric columns (assume phylogenetic columns are binary: 0 or 1)
        vec1 = row1.select_dtypes(include=[np.number]).values.flatten()
        vec2 = row2.select_dtypes(include=[np.number]).values.flatten()
        
        # Ensure both vectors are same length and valid
        if len(vec1) != len(vec2) or len(vec1) == 0:
            return np.nan
        
        intersection = np.logical_and(vec1, vec2).sum()
        union = np.logical_or(vec1, vec2).sum()
        return intersection / union if union != 0 else np.nan

    except Exception as e:
        print(f"Error comparing {g1} and {g2}: {e}")
        return np.nan

pp_sims = [phylo_similarity(*pp_pairs[i], phylo_df) for i in pp_sample]
# pn_sims = [phylo_similarity(*pn_pairs[i], phylo_df) for i in pn_sample]
# nn_sims = [phylo_similarity(*nn_pairs[i], phylo_df) for i in nn_sample]
```

```python
from tqdm import tqdm
```

```python
pp_sims = [phylo_similarity(g1, g2, phylo_df) for g1, g2 in tqdm(pp_pairs, desc="Computing PP similarities")]
```

```python
pp_df = pd.DataFrame(pp_pairs, columns=["Gene_1", "Gene_2"])
pp_df["Phylogenetic_Jaccard_Similarity"] = pp_sims

pp_df.to_csv('phylogenetic_similarity_cilia.csv', index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Combine scores into a single DataFrame
df = pd.DataFrame({
    "similarity": pp_sims + pn_sims + nn_sims,
    "pair_type": ["positive-positive"] * len(pp_sims) +
                 ["positive-negative"] * len(pn_sims) +
                 ["negative-negative"] * len(nn_sims)
})

# Plot KDE
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x="similarity", hue="pair_type", common_norm=False, fill=True, alpha=0.4, linewidth=2)

plt.title("Phylogenetic Profile Similarity Distributions (Jaccard)", fontsize=14)
plt.xlabel("Jaccard Similarity", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

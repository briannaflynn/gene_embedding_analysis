```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
reference_proteome = set()
with open("HumanRefProteome_UniProtACC_EnsGeneID.tsv") as INPUT:
    for line in INPUT:
        ensID = line.strip().split("\t")[1]
        if ensID != "N/A":
            ensID_list = ensID.split(",")
            reference_proteome.update(ensID_list)
```

```python
df = pd.read_csv("ratio_of_n_and_n_cells_cell_type.gene.tissue-celltype.ver2.tsv", sep="\t", header=0, index_col=0)
```

```python
df = df.loc[df.index.isin(reference_proteome)]
```

```python
df.shape
```

```python
# ('ENSG00000198712', 'ENSG00000198938')  0.9305114961321003  1
# MT-CO2: mitochondrially encoded cytochrome c oxidase II
# MT-CO3: mitochondrially encoded cytochrome c oxidase III
gene1 = "ENSG00000198712" # MT-C02
gene2 = "ENSG00000198938" # MT-C03
```

```python
plt.figure(figsize=(12, 5))
x_values = np.arange(1, df.shape[1] + 1)
plt.plot(x_values, df.loc[gene1].values, label=f"{gene1}", color="blue", alpha=0.7, linewidth=.7)
plt.plot(x_values, df.loc[gene2].values, label=f"{gene2}", color="orange", alpha=0.7, linewidth=.7)

plt.xlabel("[Cell type x tissue]")
plt.ylabel("Ratio of expressing cells")
plt.grid()
#plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('RF_ptm_by_bin.csv')
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='bin', y='ptm')
plt.title("PTM Score Distribution by Bin")
plt.xlabel("Probability Bin")
plt.ylabel("PTM Score")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.violinplot(data=df, x='bin', y='ptm', inner='quartile')
plt.title("PTM Score Density by Bin")
plt.xlabel("Probability Bin")
plt.ylabel("PTM Score")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
bin_means = df.groupby("bin")["ptm"].mean()
bin_std = df.groupby("bin")["ptm"].std()

plt.figure(figsize=(8, 5))
plt.errorbar(bin_means.index, bin_means.values, yerr=bin_std.values, fmt='-o', capsize=4)
plt.title("Mean PTM Score by Bin")
plt.xlabel("Probability Bin")
plt.ylabel("Mean PTM Score  SD")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='bin', y='ptm')
sns.stripplot(data=df, x='bin', y='ptm', color='black', size=2.5, jitter=0.25)
plt.title("PTM Scores by Bin with Individual Points")
plt.xlabel("Probability Bin")
plt.ylabel("PTM Score")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
